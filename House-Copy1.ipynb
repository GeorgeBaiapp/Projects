```python
import numpy
import pandas as pd

# Wtf is mode of voting;
# candidatevotes: for uncontested races, value is set to 1 in FL. Should user want to set a higher value for analysis 
'''
district: district is set to 0 for single member states. 

party and candidate: candidate - party combinations are recorded as they were on the state elections website. This
means that for states where the same candidate might appear on multiple parties, like in NY, they are recorded as
such. Therefore, for users interested in finding the primary party, run the following code: 
'''

house_dataset = pd.read_csv(
    '1976-2020-house.csv')
```

```python
house_dataset.isna().sum()

clean_dataset = house_dataset.dropna(subset=["candidate"])
clean_dataset
```

```python
column = [x for x in clean_dataset.columns if (x != "candidatevotes" and x != "party")]
kill_me = clean_dataset.groupby(column, as_index=False, dropna=False)['candidatevotes'].sum()
kill_me
# check why when not including party, it added rows...
# 30999 rows. 
```

```python
winner = kill_me.groupby(["year", "state", "district"], sort=False).apply(lambda x: x.sort_values(['candidatevotes'], ascending=False)).reset_index(drop=True)
pd.set_option('display.max_rows', 100)
winner

# This sorts the house candidatevotes by ascending value.
```

```python
true_winner = winner.groupby(["year", "state", "district"], sort=False).first()
true_winner

# These are all the winners. I think this makes sense.
```
